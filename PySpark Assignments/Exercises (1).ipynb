## WITHCOLUMN Operations

```python
# Create bonus column as 10% of salary
from pyspark.sql.functions import col
df.withColumn("bonus", col("salary") * 0.10).show()
```

```python
# Create annual_salary column
df.withColumn("annual_salary", col("salary") * 12).show()
```

```python
# Create tax column as 5% of salary
df.withColumn("tax", col("salary") * 0.05).show()
```

```python
# Create updated_salary by adding 5000
df.withColumn("updated_salary", col("salary") + 5000).show()
```

```python
# Create salary_category column: High, Medium, Low
from pyspark.sql.functions import when
df.withColumn("salary_category", 
    when(col("salary") > 80000, "High")
    .when(col("salary") > 50000, "Medium")
    .otherwise("Low")
).show()
```

```python
# Create age_group column: Young, Adult
df.withColumn("age_group",
    when(col("age") < 28, "Young")
    .otherwise("Adult")
).show()
```

```python
# Create location column by combining city and department
from pyspark.sql.functions import concat, lit
df.withColumn("location", concat(col("city"), lit("-"), col("department"))).show()
```

```python
# Create increment_salary column with 15% hike
df.withColumn("increment_salary", col("salary") * 1.15).show()
```

```python
# Create experience_status column based on joining year
from pyspark.sql.functions import year, current_date
df.withColumn("experience_status",
    when(year(current_date()) - year(col("joining_date")) > 4, "Experienced")
    .otherwise("Fresher")
).show()
```

```python
# Create name_length column using employee name
from pyspark.sql.functions import length
df.withColumn("name_length", length(col("emp_name"))).show()
```

```python
# Create is_high_salary column using condition
df.withColumn("is_high_salary",
    when(col("salary") > 70000, True)
    .otherwise(False)
).show()
```

```python
# Create joining_year column from joining_date
df.withColumn("joining_year", year(col("joining_date"))).show()
```

```python
# Create salary_after_tax column
df.withColumn("salary_after_tax", col("salary") * 0.90).show()
```

```python
# Create department_code column
from pyspark.sql.functions import substring
df.withColumn("department_code", substring(col("department"), 1, 2)).show()
```

```python
# Create double_salary column
df.withColumn("double_salary", col("salary") * 2).show()
```
